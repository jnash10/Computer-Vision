```python
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
import numpy as np
import cv2
```

```python
img = cv2.imread('touching.png', cv2.IMREAD_GRAYSCALE)
```

```python
img, img.shape
```

```python
pixel_vals = img.reshape((-1,1))
pixel_vals.shape, pixel_vals
```

```python
plt.imshow(img, cmap='gray')
```

```python
kmeans = KMeans(n_clusters=2, random_state=0, n_init=10).fit(pixel_vals)
```

```python
clustered = kmeans.labels_.reshape((166,150))
```

```python
plt.imshow(clustered, cmap='gray')
```

```python
#now trying with 3 clusters

kmeans_3 = KMeans(n_clusters=3, random_state=0, n_init=10).fit(pixel_vals)

clustered_3 = kmeans_3.labels_.reshape(img.shape)
plt.imshow(clustered_3, cmap='gray')
```
