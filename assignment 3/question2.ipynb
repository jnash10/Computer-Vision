```python
import matplotlib.pyplot as plt
import cv2
```

```python
im1 = cv2.imread('eiffel1.jpg', cv2.IMREAD_GRAYSCALE)
im2 = cv2.imread('eiffel2.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
cim1 = cv2.cornerHarris(im1,2,3,0.4)

cim2 = cv2.cornerHarris(im2,2,3,0.4)
```

```python
plt.imshow(cim1, cmap='gray')
```

```python
plt.imshow(cim2, cmap='gray')
```
