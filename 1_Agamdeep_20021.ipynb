```python
import cv2
import numpy as np
```

```python
img = cv2.imread('/home/agam/classes/cv/class_photo.jpeg')
```

```python
def show(img):
    cv2.imshow("test",img)
    cv2.waitKey(5000)
    cv2.destroyAllWindows()
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
show(gray)
```

```python
new = np.zeros(gray.shape, dtype = gray.dtype)
```

```python
show(new)
```

```python
angle_of_rotation = np.pi/6
rotation_mat = np.array([[np.cos(angle_of_rotation),np.sin(angle_of_rotation)],
                            [-np.sin(angle_of_rotation),np.cos(angle_of_rotation)]])
```

```python
rotation_mat
```

```python
for i in range(gray.shape[0]): #i iterates over rows
    for j in range(gray.shape[1]): #j iterates over columns
        new_vec = np.matmul(rotation_mat, [[i],[j]])
        new_vec = new_vec.astype(int)
        try:
            new[new_vec[0][0]][new_vec[1][0]] = gray[i][j]
        except:
            pass
    
    
```

```python
show(new)
```

```python
translate = np.zeros((gray.shape),dtype=gray.dtype)
```

```python
x_shift = 10
y_shift = 20

for i in range(gray.shape[0]):
    for j in range(gray.shape[1]):
        try:
            translate[i+x_shift][j+y_shift]=gray[i][j]
        except:
            pass
```

```python
show(translate)
```

```python
np.array(gray.shape)*5
```

```python
scale_factor = 0.5
shape = np.array(gray.shape)*scale_factor
shape = shape.astype(int)
scaled = np.zeros(shape,dtype=gray.dtype)
```

```python
for i in range(gray.shape[0]):
    for j in range(gray.shape[1]):
        try:
            scaled[int(i*scale_factor)][int(j*scale_factor)]=gray[i][j]
        except:
            pass
```

```python
show(scaled)
```
